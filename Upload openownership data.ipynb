# Look on data

```python
with open(source_file, 'r') as src:
    data0 = json.loads(src.readline())
    person_id = data0['interestedParty']['describedByPersonStatement']
    entity = data0['subject']['describedByEntityStatement']
    print(person_id, entity)
    print(json.dumps(data0, indent=2))
    step = 0
    found = 0
    while True:
        
        data = json.loads(src.readline())
        if data['statementID'] in (person_id, entity):
            print(step)
            print(json.dumps(data, indent=2))  
            if data['statementID'] == person_id:
                for name in data['names']:
                    print(name['type'])
                found += 1
            else:
                found += 10
        if not data:
            break
        if found >= 11:
            print(f"Found {found}")
            break
        step += 1;
        if step%10000 == 0:
            print(f"Step {step} done")
    
```

# Data exploration

```python
elem_sizes = {}
elem_sizes['statementID'] = 0
elem_sizes['entityType'] = 0
elem_sizes['name'] = 0
elem_sizes['jurisdiction_name'] = 0
elem_sizes['jurisdiction_code'] = 0
elem_sizes['addresses_type'] = 0
elem_sizes['addresses_address'] = 0
elem_sizes['addresses_country'] = 0
elem_sizes['personType'] = 0
elem_sizes['names_type'] = 0
elem_sizes['names_fullName'] = 0
elem_sizes['names_familyName'] = 0
elem_sizes['names_givenName'] = 0
elem_sizes['nationalities_name'] = 0
elem_sizes['nationalities_code'] = 0

with open(source_file, 'r', errors='ignore') as src:
    step = 0
    while True:      
        read_data = src.readline()
        if not read_data:
            print(f"Step {step} done")
            break
        data = json.loads(read_data)
        try:
            if elem_sizes['statementID'] < int(data['statementID']):
                elem_sizes['statementID'] = int(data['statementID'])
            if data['statementType'] in ('entityStatement', 'personStatement'):
                for address in data['addresses']:
                    if elem_sizes['addresses_type'] < len(address['type']):
                        elem_sizes['addresses_type'] = len(address['type'])
                    if elem_sizes['addresses_address'] < len(address['address']):
                        elem_sizes['addresses_address'] = len(address['address'])
                    if elem_sizes['addresses_country'] < len(address['country']):
                        elem_sizes['addresses_country'] = len(address['country'])
            if data['statementType'] == 'entityStatement':
                if elem_sizes['entityType'] < len(data['entityType']):
                    elem_sizes['entityType'] = len(data['entityType'])  
                if elem_sizes['name'] < len(data['name']):
                    elem_sizes['name'] = len(data['name']) 
                if elem_sizes['jurisdiction_name'] < len(data['incorporatedInJurisdiction']['name']):
                    elem_sizes['jurisdiction_name'] = len(data['incorporatedInJurisdiction']['name'])
                if elem_sizes['jurisdiction_code'] < len(data['incorporatedInJurisdiction']['code']):
                    elem_sizes['jurisdiction_code'] = len(data['incorporatedInJurisdiction']['code'])
            if data['statementType'] == 'personStatement':
                if elem_sizes['personType'] < len(data['personType']):
                    elem_sizes['personType'] = len(data['personType']) 
                for name in data['names']:
                    if elem_sizes['names_type'] < len(name['type']):
                        elem_sizes['names_type'] = len(name['type'])
                    if elem_sizes['names_fullName'] < len(name['fullName']):
                        elem_sizes['names_fullName'] = len(name['fullName'])
                    if elem_sizes['names_familyName'] < len(name['familyName']):
                        elem_sizes['names_familyName'] = len(name['familyName'])
                    if elem_sizes['names_givenName'] < len(name['givenName']):
                        elem_sizes['names_givenName'] = len(name['givenName'])
                for nation in data['nationalities']:
                    if elem_sizes['nationalities_name'] < len(nation['name']):
                        elem_sizes['nationalities_name'] = len(nation['name'])
                    if elem_sizes['nationalities_code'] < len(nation['code']):
                        elem_sizes['nationalities_code'] = len(nation['code'])  
        except KeyError:
            pass
        step += 1;
        if step%1000000 == 0:
            print(f"Step {step} done")
    
print(elem_sizes)
```

```python
statement_types = {}

with open(source_file, 'r', errors='ignore') as src:
    step = 0
    while True:      
        read_data = src.readline()
        if not read_data:
            print(f"Step {step} done")
            break
        data = json.loads(read_data)
        try:
            statement_types[data['statementType']] += 1
        except KeyError:
            statement_types[data['statementType']] = 1
        step += 1;
        if step%1000000 == 0:
            print(f"Step {step} done")
    
print(statement_types)
```

```python
conn = psycopg.connect(dbname="postgres", user="postgres", password="deaththekid")
try:
    cur = conn.cursor()
    try:
        cur.execute("""
                    TRUNCATE TABLE raw_data.openownership_relationships;
                    TRUNCATE TABLE raw_data.openownership_addresses;
                    TRUNCATE TABLE raw_data.openownership_entities;
                    TRUNCATE TABLE raw_data.openownership_persons;
                    TRUNCATE TABLE raw_data.openownership_person_names;
                    TRUNCATE TABLE raw_data.openownership_person_nationalities;
                    """)
        conn.commit()
    except:
        cur.close()
except:
    conn.close()
    
conn = psycopg.connect(dbname="postgres", user="postgres", password="deaththekid")
try:
    cur = conn.cursor()
    try:
        with open(source_file, 'r', errors='ignore') as src:
            step = 0
            while True:  
                try:
                    read_data = src.readline()
                except UnicodeDecodeError:
                    continue
                if not read_data:
                    break
                try:
                    data = json.loads(read_data)
                except:
                    continue

                try:
                    if data['statementType'] == 'ownershipOrControlStatement':
                        try:
                            subject = list(data['subject'].values())[0] 
                        except KeyError as e:
                            subject = ''
                        try:
                            interestedParty = list(data['interestedParty'].values())[0]
                        except KeyError as e:
                            interestedParty = ''
                        
                        cur.execute("""INSERT INTO raw_data.openownership_relationships
                                    (statementID, statementDate, isComponent, subject_statement_id, interest_statement_id, publicationDate)
                                    VALUES (%s, %s, %s, %s, %s, %s)"""
                                    , (data['statementID'], data['statementDate'], data['isComponent']
                                       , subject, interestedParty
                                       , data['publicationDetails']['publicationDate']))
                        
                    else:
                        try:
                            for address in data['addresses']:
                                try:
                                    country = address['country']
                                except KeyError as e:
                                    country = ''
                                cur.execute("""INSERT INTO raw_data.openownership_addresses
                                            (statementID, statementType, address, country, type)
                                            VALUES (%s, %s, %s, %s, %s)"""
                                            , (data['statementID'], data['statementType'], address['address'], country, address['type']))
                        except KeyError as e:
                            pass
                        if data['statementType'] == 'entityStatement':
                            try:
                                foundingDate = data['foundingDate']
                            except KeyError as e:
                                foundingDate = ''
                            try:
                                name = data['name']
                            except KeyError as e:
                                name = ''
                            try:
                                incorporatedInJurisdiction = data['incorporatedInJurisdiction']
                            except KeyError as e:
                                incorporatedInJurisdiction = {'name': '', 'code': ''}
                            cur.execute("""INSERT INTO raw_data.openownership_entities
                                        (statementID, statementDate, isComponent, entityType, name, jurisdiction, jurisdiction_code, foundingDate, publicationDate)
                                        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)"""
                                        , (data['statementID'], data['statementDate'], data['isComponent'], data['entityType']
                                           , name, incorporatedInJurisdiction['name'], incorporatedInJurisdiction['code']
                                           , foundingDate, data['publicationDetails']['publicationDate']))

                        if data['statementType'] == 'personStatement':
                            try:
                                birthDate = data['birthDate']
                            except KeyError as e:
                                birthDate = ''
                            cur.execute("""INSERT INTO raw_data.openownership_persons
                                        (statementID, statementDate, isComponent, personType, birthDate, publicationDate)
                                        VALUES (%s, %s, %s, %s, %s, %s)"""
                                        , (data['statementID'], data['statementDate'], data['isComponent'], data['personType']
                                           , birthDate, data['publicationDetails']['publicationDate']))
                            try:
                                for name in data['names']:
                                    cur.execute("""INSERT INTO raw_data.openownership_person_names
                                                (statementID, statementType, type, fullName, familyName, givenName)
                                                VALUES (%s, %s, %s, %s, %s, %s)"""
                                                , (data['statementID'], data['statementType'], name['type'], name['fullName'], name['familyName'], name['givenName']))
                            except KeyError as e:
                                pass
                            try:
                                for nat in data['nationalities']:
                                    cur.execute("""INSERT INTO raw_data.openownership_person_nationalities
                                                (statementID, statementType, nationality_name, nationality_code)
                                                VALUES (%s, %s, %s, %s)"""
                                                , (data['statementID'], data['statementType'], nat['name'], nat['code']))
                            except KeyError as e:
                                pass
                except KeyError as e: 
                    print('Error: ', data['statementID'],  data['statementType'])
                    print(json.dumps(data, indent=2)) 
                    raise e
                step += 1;
                if step%1000000 == 0:
                    conn.commit()
                    print(f"Step {step} done")

        conn.commit()        
        cur.close()
        conn.close()
    except Error as e:
        cur.close()
        raise e        
except Error as e:
    conn.close()
    raise e 
```
